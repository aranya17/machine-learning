# **Models**

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
x = df.drop(columns = 'quality')
y = df['quality']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

print("X train :", x_train.shape)
print("X test :", x_test.shape)
print("Y train :", y_train.shape)
print("Y test :", y_test.shape)
```

```python
model = {}
```

**Logistic Regression**

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred1 = lr.predict(x_test)
print("Accuracy :", accuracy_score(y_test, y_pred1).round(3))
print("F1 Score :", f1_score(y_test, y_pred1, average='weighted').round(3))
print("Classification Report :\n", classification_report(y_test, y_pred1))

model_name = 'Logistic Regression'
model[model_name] = [accuracy_score(y_test, y_pred1), f1_score(y_test, y_pred1, average='weighted')]
```

**Decision Tree**

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred2 = dt.predict(x_test)
print("Accuracy :", accuracy_score(y_test, y_pred2).round(3))
print("F1 Score :", f1_score(y_test, y_pred2, average='weighted').round(3))
print("Classification Report :\n", classification_report(y_test, y_pred2))

model_name = 'Decision Tree'
model[model_name] = [accuracy_score(y_test, y_pred2), f1_score(y_test, y_pred2, average='weighted')]
```

**Random Forest**

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred3 = rf.predict(x_test)
print("Accuracy :", accuracy_score(y_test, y_pred3).round(3))
print("F1 Score :", f1_score(y_test, y_pred3, average='weighted').round(3))
print("Classification Report :\n", classification_report(y_test, y_pred3))

model_name = 'Random Forest'
model[model_name] = [accuracy_score(y_test, y_pred3), f1_score(y_test, y_pred3, average='weighted')]
```

**KNN**

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred4 = knn.predict(x_test)
print("Accuracy :", accuracy_score(y_test, y_pred4).round(3))
print("F1 Score :", f1_score(y_test, y_pred4, average='weighted').round(3))
print("Classification Report :\n", classification_report(y_test, y_pred4))

model_name = 'KNN'
model[model_name] = [accuracy_score(y_test, y_pred4), f1_score(y_test, y_pred4, average='weighted')]
```

**SVC**

```python
svc = SVC()
svc.fit(x_train, y_train)
y_pred5 = svc.predict(x_test)
print("Accuracy :", accuracy_score(y_test, y_pred5).round(3))
print("F1 Score :", f1_score(y_test, y_pred5, average='weighted').round(3))
print("Classification Report :\n", classification_report(y_test, y_pred5))

model_name = 'SVC'
model[model_name] = [accuracy_score(y_test, y_pred5), f1_score(y_test, y_pred5, average='weighted')]
```

**Model Comparison**

```python
df1 = pd.DataFrame.from_dict(model).T
df1.columns = ['Accuracy', 'F1 Score']
df1 = df1.sort_values(by='Accuracy', ascending=True)
df1.style.background_gradient(cmap='Blues')
```
